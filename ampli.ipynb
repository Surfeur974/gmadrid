### OTA 5 transistors (paire différentielle NMOS M1/M2, charge active PMOS M3/M4, source M5)
 

| Caractéristique | Symbole | Expression / Valeur typique | Commentaire |
|-----------------|---------|------------------------------|-------------|
| Transconductance équivalente | $G_m$ | $G_m = g_{m1}$ | Paire différentielle NMOS d’entrée |
| Résistance de sortie | $R_{out}$ | $R_{out} = r_{o2} \parallel r_{o4}$ | Sortie simple |
| Gain en tension DC | $A_{v0}$ | $A_{v0} = g_{m1}\,(r_{o2} \parallel r_{o4})$ | ~30–50 dB |
| Pôle dominant | $f_{p1}$ | $f_{p1} = \dfrac{1}{2\pi R_{out} C_L}$ | $C_L$ : capacité de charge totale consider CDD of both output MOS |
| Produit gain‑bande (GBW) | $GBW$ | $GBW = \dfrac{g_{m1}}{2\pi C_L}$ | Égal à $f_u$ |
| Pôles non‑dominants | $f_{p2}$ | $\sim\displaystyle \frac{1}{2\pi\,g_{m1}C_{gs1}}$ | Négligeable en pratique |
| Marge de phase | $\phi_m$ | $\approx 90^\circ$ (système à un pôle dominant) | |
| Vitesse de balayage (Slew Rate) | $SR$ | $SR = \dfrac{I_{SS}}{C_L}$ | $I_{SS}$ : courant de queue ($M_5$) |
| Plage de mode commun d’entrée (ICMR) | $V_{IC,\min}$ / $V_{IC,\max}$ | $V_{IC,\min} = V_{SDsat,1} + V_{SDsat,5} +V_{TH1}$<br> $V_{IC,\max} = V_{DD}-V_{SDsat,3}$ | |
| Excursion de sortie | $V_{out,\min}$ / $V_{out,\max}$ | $V_{out,\min} = V_{DSsat,1} + V_{SDsat,5} $<br>$V_{out,\max} = V_{DD} - V_{SDsat,3}$ |  |
| CMRR | $CMRR$ | $CMRR \approx \dfrac{2\,g_{m1}\,g_{m3}\,r_{o5}}{g_{ds2}+g_{ds4}}$ | ~40–60 dB |
| PSRR | $PSRR$ | Faible (alimentation module les miroirs) | |
| Tension d’offset d’entrée | $V_{OS}$ | Aléatoire, due aux désadaptations ($\Delta V_{TH}$, $\Delta W/L$) de $M_1,M_2$ et $M_3,M_4$ | Appairage soigné requis |
| Bruit ramené à l’entrée (densité spectrale) | $\overline{v_{n,\text{in}}^2}$ | $\displaystyle \overline{v_{n,\text{in}}^2} = 8kT\gamma\left(\frac{1}{g_{m1}}+\frac{g_{m3}}{g_{m1}^2}\right) + \frac{KF_n}{(WL)_1 C_{ox} f} + \frac{KF_p}{(WL)_3 C_{ox} f}\left(\frac{g_{m3}}{g_{m1}}\right)^2$ | Bruit de $M_5$ rejeté en mode commun |

### OTA deux étages avec compensation Miller (1er étage 5T, 2e étage NMOS M6/charge M7, $C_C$, $R_z$ optionnel)


| Caractéristique | Symbole | Expression / Valeur typique | Commentaire |
|-----------------|---------|------------------------------|-------------|
| Transconductance équivalente | $G_m$ | $G_m = g_{m1}$ (1er étage) | Le second étage amplifie la tension |
| Résistance de sortie | $R_{out}$ | $R_{out} = r_{o6} \parallel r_{o7}$ | Sortie de l’étage common‑source |
| Gain en tension DC | $A_{v0}$ | $A_{v0} = A_{v1}A_{v2} = \bigl[g_{m1}(r_{o2}\|r_{o4})\bigr]\,\bigl[g_{m6}(r_{o6}\|r_{o7})\bigr]$ | ~60–90 dB |
| Pôle dominant | $f_{p1}$ | $f_{p1} \approx \dfrac{1}{2\pi (r_{o2}\|r_{o4}) A_{v2} C_c}$ | Effet Miller de $C_c$ |
| Produit gain‑bande (GBW) | $GBW$ | $GBW \approx \dfrac{g_{m1}}{2\pi C_c}$ | Indépendant de $C_L$ |
| Pôle non‑dominant | $f_{p2}$ | $f_{p2} \approx \dfrac{g_{m6}}{2\pi C_L}$ | Associé au nœud de sortie |
| Zéros | $f_z$, $f_{z,R}$ | $f_z = \dfrac{g_{m6}}{2\pi C_c}$ (RHP) ; $f_{z,R} = \dfrac{1}{2\pi C_c (1/g_{m6} - R_z)}$ | Annulation avec $R_z = 1/g_{m6}$ |
| Marge de phase | $\phi_m$ | $\phi_m = 90^\circ - \arctan(GBW/f_{p2}) - \arctan(GBW/\lvert f_z\rvert)$ | Vise $\ge 60^\circ$ |
| Slew Rate | $SR$ | $SR = \min\left(\dfrac{I_{SS}}{C_c},\; \dfrac{I_{D6}}{C_c+C_L}\right)$ | $I_{SS}$ : courant de queue ; $I_{D6}$ : polarisation de $M_6$ |
| Plage de mode commun d’entrée (ICMR) | $V_{IC,\min}$ / $V_{IC,\max}$ | $V_{IC,\min} = V_{SDsat,1} + V_{SDsat,5} +V_{TH1}$<br> $V_{IC,\max} = V_{DD}-V_{SDsat,3}$ | |
| Excursion de sortie | $V_{out,\min}$ / $V_{out,\max}$ | $\displaystyle [V_{SS}+V_{DSsat,7},\;V_{DD}-V_{SDsat,6}]$ | |
| CMRR | $CMRR$ | $\approx g_{m1}r_{o5}$ | |
| PSRR | $PSRR$ | Faible en HF, modéré en BF | |
| Tension d’offset | $V_{OS}$ | Principalement due au 1er étage | |
| Bruit ramené à l’entrée | $\overline{v_{n,\text{in}}^2}$ | TO ADD | Le 1er étage domine |

# OTA telescopic cascode  
  

| Caractéristique | Symbole | Expression / Valeur typique | Commentaire |
|-----------------|---------|------------------------------|-------------|
| Transconductance équivalente | $G_m$ | $G_m = g_{m1}$ | Paire d’entrée NMOS |
| Résistance de sortie | $R_{out}$ | $R_{out,TC\_OTA} = \frac{1}{\displaystyle\frac{g_{ds4}\,g_{ds9}}{g_{m9}} \;+\; \frac{g_{m1}\,g_{ds2}\,g_{ds7}}{g_{m2}\,g_{m7}}} = \frac{1}{2}\,\frac{g_m}{g_{ds}^2}$ | Très élevée grâce aux cascodes |
| Gain en tension DC | $A_{v0}$ | $A_{v0} \approx \frac{1}{2}\,g_m^2 r_o^2$ | ~60–80 dB |
| Pôle dominant | $f_{p1}$ | $f_{p1} = \dfrac{1}{2\pi R_{out} C_L}$ | |
| GBW | $GBW$ | $GBW = \dfrac{g_{m1}}{2\pi C_L}$ | |
| Pôles non‑dominants | $f_{p2}$ | $f_{p2} \approx \dfrac{g_{m3}}{2\pi C_x}$ (cascode NMOS) et symétrique PMOS | $C_x = C_{db1}+C_{gd1}+C_{sb3}+C_{gs3}$ |
| Marge de phase | $\phi_m$ | Bonne si $f_{p2,\min} > 3\times GBW$ | |
| Slew Rate | $SR$ | $SR = \dfrac{I_{SS}}{C_L}$ | $I_{SS}$ : courant de queue ($M_9$) |
| ICMR | $V_{IC,\min}$ / $V_{IC,\max}$ | $V_{IC,\min} = V_{DS1,\text{sat}} + V_{DS5,\text{sat}} + V_{TH}$<br>$V_{IC,\max} = V_{DD} -  V_{DS8,\text{sat}} - V_{DS3,\text{sat}}$ | Limitée côté haut par l’empilement cascode |
| Excursion de sortie | $V_{out,\min}$ / $V_{out,\max}$ | $V_{out,\min} = V_{DS5,\text{sat}}+V_{DS2,\text{sat}}+V_{DS7,\text{sat}}$ <br>$V_{out,\max} = V_{DD} -  V_{DS4,\text{sat}} - V_{DS9,\text{sat}}$ | Trois NMOS cascodés fixent la limite basse |
| CMRR | $CMRR$ | Très bon ($>80$ dB) | Impédance de queue élevée + gain cascode |
| PSRR | $PSRR$ | Bon en BF, dégradé en HF | |
| Tension d’offset | $V_{OS}$ | Dominée par $M_1,M_2$ et le miroir de charge PMOS | |
| Bruit ramené à l’entrée | $\overline{v_{n,\text{in}}^2}$ | $\displaystyle \overline{v_{n,\text{in}}^2} \approx 8kT\gamma\left(\frac{1}{g_{m1}}+\frac{g_{m7}}{g_{m1}^2}\right) + \frac{KF_n}{(WL)_1 C_{ox} f} + \frac{KF_p}{(WL)_7 C_{ox} f}\left(\frac{g_{m7}}{g_{m1}}\right)^2$ | Cascodes (M3–M6) négligeables (atténuation $g_m r_o$) |

# Folded Cascode  
 

| Caractéristique | Symbole | Expression / Valeur typique | Commentaire |
|-----------------|---------|------------------------------|-------------|
| Transconductance équivalente | $G_m$ | $G_m = g_{m1}$ | Paire d’entrée NMOS |
| Résistance de sortie | $R_{out}$ | $R_{out} \approx \frac{1}{2}\,g_m r_o^2$ | Comparable au télescopique |
| Gain en tension DC | $A_{v0}$ | $A_{v0} \approx \frac{1}{2}\,g_m^2 r_o^2$ | ~60–80 dB |
| Pôle dominant | $f_{p1}$ | $f_{p1} = \dfrac{1}{2\pi R_{out} C_L}$ | |
| GBW | $GBW$ | $GBW = \dfrac{g_{m1}}{2\pi C_L}$ | |
| Pôles non‑dominants | $f_{p,\text{fold}}, f_{p,\text{mir}}$ | $f_{p,\text{fold}} \approx \dfrac{g_{m,\text{fold}}}{2\pi C_x}$ (nœud de repliement)<br>$f_{p,\text{mir}}$ lié au miroir cascode NMOS | |
| Marge de phase | $\phi_m$ | Bonne si les pôles non‑dominants sont $> 3\times GBW$ | |
| Slew Rate | $SR$ | $SR = \min\left(\dfrac{I_{SS}}{C_L},\; \dfrac{I_{\text{fold}}}{C_L}\right)$ | $I_{\text{fold}}$ : courant de la branche de repliement |
| ICMR | $V_{IC,\min}$ / $V_{IC,\max}$ | $V_{IC,\min} = V_{DS1,\text{sat}} + V_{DS5,\text{sat}} + V_{TH}$<br>$V_{IC,\max} = V_{DD} -  V_{DS8,\text{sat}} + V_{TH}$ | Très large, proche de $V_{DD}$ |
| Excursion de sortie | $V_{out,\min}$ / $V_{out,\max}$ | $V_{out,\min} = V_{DS11,\text{sat}} + V_{DS7,\text{sat}}$<br>$V_{out,\max} = V_{DD} - V_{DS9,\text{sat}} - V_{DS4,\text{sat}}$ | Symétrique, meilleure que le télescopique |
| CMRR | $CMRR$ | Très bon ($>80$ dB) | |
| PSRR | $PSRR$ | Bon en BF, dégradé en HF | |
| Tension d’offset | $V_{OS}$ | Due à $M_1,M_2$, aux sources de repliement et au miroir de charge | Appairage multiple nécessaire |
| Bruit ramené à l’entrée | $\overline{v_{n,\text{in}}^2}$ | $\displaystyle \overline{v_{n,\text{in}}^2} \approx 8kT\gamma\left(\frac{1}{g_{m1}}+\frac{g_{m,\text{fold}}}{g_{m1}^2}+\frac{g_{m,\text{mir}}}{g_{m1}^2}\right) + \frac{KF_n}{(WL)_1 C_{ox} f} + \frac{KF_p}{(WL)_{\text{fold}} C_{ox} f}\left(\frac{g_{m,\text{fold}}}{g_{m1}}\right)^2 + \frac{KF_n}{(WL)_{\text{mir}} C_{ox} f}\left(\frac{g_{m,\text{mir}}}{g_{m1}}\right)^2$ | |

# Symmetrical OTA  
 

| Caractéristique | Symbole | Expression / Valeur typique | Commentaire |
|-----------------|---------|------------------------------|-------------|
| Transconductance équivalente | $G_m$ | $G_m = K \cdot g_{m1}$ | $K$ = rapport des miroirs de courant ($(W/L)_{M4}/(W/L)_{M3}$) |
| Résistance de sortie | $R_{out}$ | $R_{out} = r_{o,\text{N}} \parallel r_{o,\text{P}}$ | Sortie simple |
| Gain en tension DC | $A_{v0}$ | $A_{v0} = K\,g_{m1}\,(r_{o,\text{N}} \parallel r_{o,\text{P}})$ | Réglable, ~40–70 dB |
| Pôle dominant | $f_{p1}$ | $f_{p1} = \dfrac{1}{2\pi R_{out} C_L}$ | |
| GBW | $GBW$ | $GBW = \dfrac{K\,g_{m1}}{2\pi C_L}$ | Proportionnel à $K$ |
| Pôles non‑dominants | $f_{p2}$ | Générés par les miroirs de courant | Peuvent dégrader la stabilité |
| Marge de phase | $\phi_m$ | Variable selon les pôles/zeros des miroirs | Analyse de stabilité nécessaire |
| Slew Rate | $SR$ | $SR = K * \dfrac{I_{SS}}{C_L}$ | Limité par le courant de queue |
| Plage de mode commun d’entrée (ICMR) | $V_{IC,\min}$ / $V_{IC,\max}$ | $V_{IC,\min} = V_{SDsat,1} + V_{TH1}$<br> $V_{IC,\max} = V_{DD}-V_{SDsat,3}+V_{TH1}$ | |
| Excursion de sortie | $V_{out,\min}$ / $V_{out,\max}$ | $V_{out,\min} = V_{DS8,\text{sat}}$<br>$V_{out,\max} = V_{DD} - V_{DS6,\text{sat}}$ | Symétrique, meilleure que le télescopique |
| CMRR | $CMRR$ | Modéré (désadaptations des miroirs) | |
| PSRR | $PSRR$ | Faible | |
| Tension d’offset | $V_{OS}$ | Augmentée par les désadaptations cumulées des miroirs | |
| Bruit ramené à l’entrée | $\overline{v_{n,\text{in}}^2}$ | $\displaystyle \overline{v_{n,\text{in}}^2} = 8kT\gamma\left(\frac{1}{g_{m1}}+\frac{g_{m3}}{g_{m1}^2}+\frac{1}{K^2}\frac{g_{m,\text{out}}}{g_{m1}^2}\right) + \frac{KF_n}{(WL)_1 C_{ox} f} + \frac{KF_p}{(WL)_3 C_{ox} f}\left(\frac{g_{m3}}{g_{m1}}\right)^2 + \frac{1}{K^2}\frac{KF_n}{(WL)_{\text{out}} C_{ox} f}\left(\frac{g_{m,\text{out}}}{g_{m1}}\right)^2$ | Le bruit du miroir de sortie est atténué par $1/K^2$ |
